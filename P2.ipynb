```python
def pl_gt_2times_pw():
    pl = ds['Petal.Length']
    pw = ds['Petal.Width']
    
    for i in range(pl.shape[0]):
        if pl[i] < 2*pw[i]:
            print('incorrect values: ')
            print("petal width: ", pw[i])
            print("petal length: ", pl[i])
```
